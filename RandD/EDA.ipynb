```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data/reddit_depression_suicidewatch.csv')
df.head()
```

```python
sw = 0
dep = 0
for row in df.iterrows():
    # print(row[1][1])
    if row[1][1] == 'depression':
        dep += 1
    else:
        sw += 1
print(dep)
print(sw)
```

```python
Country = ['depression', 'suicide watch']
GDP_Per_Capita = [dep, sw]

plt.bar(Country, GDP_Per_Capita)
plt.title('Country Vs GDP Per Capita')
plt.xlabel('Country')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
df['text_length'] = df['text'].str.len()
df.to_csv('data/reddit_unencode_length.csv')
plt.figure(figsize=(12.8,6))
sns.distplot(df['text_length']).set_title('text length distribution');
```

```python
quantile_95 = df['text_length'].quantile(0.95)
df_95 = df[df['text_length'] < quantile_95]
```
